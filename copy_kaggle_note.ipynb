# 月次売上

月次のパターンをyear_monthとする

```python
# train.year_monthにて、列(Series型)ごとにユニークな値を確認(年月の組み合わせは22パターンと確認できる)
train.year_month.unique()
```

```python
train.year_month.nunique()
```

```python
import time
# 複数のリストの直積（デカルト積）を生成するためのライブラリ
from itertools import product
ts = time.time()

# 訓練データに存在する、(年月番号,店ID,商品ID)の全組み合わせを列挙した行列を生成していく
# 最終的にmatrixを学習モデルの訓練データとする
matrix = []
for i in range(train.year_month.nunique()):
    # 変数salesにdate_block_num=iの行列(表)データを代入する
    sales = train[train.year_month==i]
    # trainデータに存在する、(年月番号,店ID,商品ID)の全組み合わせを列挙した行列を追加していく
    matrix.append(np.array(list(product([i], sales.店舗ID.unique(), sales.商品ID.unique())), dtype='int16'))


# 列名を改めて設定してmatrixを更新
cols = ['year_month','店舗ID','商品ID']
matrix = pd.DataFrame(np.vstack(matrix), columns=cols)
# .astype(~~~):各特徴量を~~~でキャスト
matrix['year_month'] = matrix['year_month'].astype(np.int8)
matrix['店舗ID'] = matrix['店舗ID'].astype(np.int8)
matrix['商品ID'] = matrix['商品ID'].astype(np.int16)
# colsをソート対象とする、inplace=Trueでオブジェクトをそのまま更新
matrix.sort_values(cols,inplace=True)
time.time() - ts
# 20.83844780921936
```

```python
matrix.shape
```

```python
# trainデータに revenue(その日の収支合計) を追加します
train['revenue'] = train['商品価格'] *  train['売上個数']
train.head()
```

```python
# trainデータにて、'date_block_num','shop_id','item_id'でGROUP化したDataFrameGroupByオブジェクトに対して、'item_cnt_day'を集計します
group = train.groupby(['year_month','店舗ID','商品ID']).agg({'売上個数': ['sum']})
# 列名の更新
group.columns = ['月次売上個数']
# DataFrameGroupBy -> DataFrame に変換
group.reset_index(inplace=True)
group.head()
```

```python
# DataFrame同士でcolsを条件に左結合する
matrix = pd.merge(matrix, group, on=cols, how='left')
# item_cnt_monthの前処理
matrix['月次売上個数'] = (matrix['月次売上個数']
                                .fillna(0) # 0で穴埋めする
                                .astype(np.float16)) # 型のキャスト
```

```python
def fit_LabelEncoding(categorical_columns, training_df, Today_df):
    
    for column in tqdm(categorical_columns):
        df = Today_df.copy()
        le_column = LabelEncoder().fit(training_df[column].fillna('Na'))
        mask = df[column].isin(le_column.classes_)
        column_new_data = df[column].mask(mask).unique()
        le_column.classes_ = np.concatenate([le_column.classes_, column_new_data])
        df[column] = le_column.transform(df[column])

    return df
```
